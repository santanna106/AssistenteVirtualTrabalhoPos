```python
from pycaret.datasets import get_data
dados = get_data('insurance')
```

```python
dados.describe()
```

```python
from pycaret.regression import *
```

```python
class_setup = setup(data = dados,
                    target = 'charges',
                    normalize = True,
                    n_jobs = -1)
```

```python
modelo = compare_models()
```

```python
modelo_tunado = tune_model(modelo, optimize = 'RMSE')
```

```python
modelo_final = finalize_model(modelo_tunado)
```

```python
save_model(modelo_final, 'modelo-previsao-custos-seguro')
```
